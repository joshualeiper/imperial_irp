### Custom Comment Builder

use a custom functon to build your own comments

```python
def generate_custom_comment(title, parameters, notes):
    comment = f"## {title}\n"
    for key, value in parameters.items():
        comment += f"# {key}: {value}\n"
    comment += f"# Notes: {notes}\n"
    return comment

parameters = {"param1": 10, "param2": 20}
comment = generate_custom_comment("Calculation Parameters", parameters, "Ensure all parameters are set correctly.")
print(comment)
```

### IO

Using StringIO to create virtual files. 

I think this is the best approach for the project. I'll try experimenting with custom functions and classes to build custom input files for PHREEQ on the fly.

```python
import io

# Using StringIO with a context manager
with io.StringIO() as virtual_file:
    virtual_file.write("This is a line of text.\n")
    virtual_file.seek(0)
    content = virtual_file.read()
print(content)
```
